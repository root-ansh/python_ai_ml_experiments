### Perceptron
Perceptron is a single layer neural network and a multi-layer perceptron is called Neural Networks.Thus , perceptrons are the base of deep learning.It is basically a linear classifier , as described previously :  
- It reads sample data (aka **inputs**) and creates a model.
- It then reads any new input and multiplies it with a **weight** and adds/subtacts the *bias* . these weights are used as edgelabels and bias is often entered as another input(eg input node will be 1 and edgeweight will be -18) or simply as nodelabel. all these values are ***summed***.  
- based on the value generated in the last step, it will predict a value, 0 or 1 based on certain condition ( eg , if Val(from last step) >2 , the 1 else 0) this condition checker function is called **Activation Function**.
- based on this prediction , new values are plotted  
, thus the perceptron is more like the image on right.

**Here is a small working example of a model in a quiz format:**(check model_verifier() function)

```python
# A popo(plain old py object) class containing data, its correct output and its prediction output
class testing_data:
    inp1 = 0
    inp2 = 0
    predictive_result = False
    correct_result = False

    def __init__(self, inp1, inp2, correct_result=False):
        self.inp1 = inp1
        self.inp2 = inp2
        self.correct_result = correct_result
    def set_prediction(self, prediction:bool):
        self.predictive_result = prediction
        
    def __str__(self):
        i1 = "inp1=" + str(self.inp1 != 0)      +"(%i)"%self.inp1
        i2 = "inp2=" + str(self.inp2 != 0)      +"(%i)"%self.inp2
        correct_result = "correct results=" + str(self.correct_result)
        predictive = "predictive results=" + str(self.predictive_result)
        
        s = i1 + '  ;  ' + i2 + '  ;  ' + correct_result + '  ;  ' + predictive
        return s

    def __repr__(self):
        return str(self)
```

```python
# just a helper function for main() function, not necessary
def is_correct_submission(inp: testing_data):
    return 'yes' if inp.correct_result == inp.predictive_result else 'no'
```

```python
#main() function
def model_verifier(w1,w2,b):
    # model verifier function
    test_inputs = [testing_data(0, 0, False), testing_data(0, 1, False), testing_data(1, 0, False), testing_data(1, 1, True)]
    for x in range(len(test_inputs)):
        pair = test_inputs[x]
        
        #------------------------model--------------------------------------
        predictor_sum = (w1 * pair.inp1) + (w2 * pair.inp2) + b     #Linear Equation Function   #(TODO: set value of w1,w2,b)
        prediction = predictor_sum >= 0                             #Activaor Function(step function)
        #--------------------------------------------------------------------
        
        pair.set_prediction(prediction)
    
    for pair in test_inputs:
        print(pair,"\t;\tis correct prediction:  "+is_correct_submission(pair))
#------------------------------

# todo:Run this function you will find that it has a column saying "is correct prediction?". set value of constants(w1,w2,b) in 
# linear equation function such that predictions are accurate for new datas and equal to actual results for pre analysed data)
w1=0
w2=0
b =0
model_verifier(w1,w2,b)
```

```python
#                           Multiplier      
#       Inp1    :   0011    *   1(w1)       =   0       0       1       1   
#       Inp2    :   0101    *   1(w2)       =   0       1       0       1
#       Bias    :   1111    *  -2(b )       =  -2      -2      -2      -2   
#---------------------------------------------------------------------------
#                                              -2      -1      -1       0  
#---------------------------------------------------------------------------
# expected=gate results for :inp1 AND inp2  =  -ve     -ve     -ve     +ve   
model_verifier(1,1,-2)
```
