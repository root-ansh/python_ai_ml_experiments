```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import scipy.stats as stats

print(np.version)
```

```python
import statsmodels.api as sm  
```

```python
dframe = sm.datasets.sunspots.load_pandas().data # loading a sample dataset
print(type(dframe))
dframe.head()
```

```python
dframe.index = pd.Index(sm.tsa.datetools.dates_from_range('1700','2008'))
del dframe['YEAR']
dframe
```

```python
plt.figure(figsize=[13,6])
plt.plot(dframe)
```

```python
dynamic_mean = dframe.rolling(window=12, center=False).mean()
dynamic_var = dframe.rolling(window=12, center=False).var()
dynamic_std = dframe.rolling(window=12, center=False).std()

fig = plt.subplots(figsize=[13, 6])

plt.plot(dframe, label='original')
plt.plot(dynamic_mean, label='mean')
# plt.plot(dynamic_var, label='variance')
plt.plot(dynamic_std, label='std')

plt.legend(loc='best')
# plt.plot(dynamic_mean)

# it also has everything increasing / decreasing but it has been  marked stationary. i don't know why?
```

```python
fig = plt.figure(figsize=[13,8])
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(dframe.values.squeeze(),lags=40,ax=ax1)
```

```python
fig = plt.figure(figsize=[13,8])
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_acf(dframe,lags=40,ax=ax2)
```

```python
arma_model = sm.tsa.ARMA(dframe, (3, 0)).fit(disp=False)

figure =plt.figure(figsize=[13,8])
plt.plot(dframe, label='original')
plt.plot(arma_model.fittedvalues, label='armamodel')
plt.legend(loc='best')

# we somehow created a 'model that fits our data' , and that's cool i guess? 
# Basically what i can see on screen is a weird blue graph almost completely covered by an orange graph
# and this means the ARMA model is very good fit for our data.
# Now we will find the 'residue' which is something related to difference between our data and model
```

```python
fig = plt.figure(figsize=[13,8])
ax = fig.add_subplot(111)
ax = arma_model.resid.plot(ax=ax)
```

```python
arma_residue = arma_model.resid

dynamic_mean = arma_residue.rolling(window=12, center=False).mean()
# dynamic_var = arma_residue.rolling(window=12, center=False).var()
dynamic_std = arma_residue.rolling(window=12, center=False).std()

fig = plt.subplots(figsize=[13, 6])

plt.plot(arma_residue, label='original')
plt.plot(dynamic_mean, label='mean')
# plt.plot(dynamic_var, label='variance')
plt.plot(dynamic_std, label='std')

plt.legend(loc='best')
# plt.plot(dynamic_mean)
```
