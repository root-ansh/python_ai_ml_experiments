## Data Cleaning

### Cleaning the price data

 Find out the number of values which are not numeric


```python
automobile['price'].str.isnumeric().value_counts()
```

```python
automobile['price'].loc[automobile['price'].str.isnumeric() == False]
```


Setting the missing value to mean of price and convert the datatype to integer

```python
price = automobile['price'].loc[automobile['price'] != '?']
```

```python
type(price)
```

```python
pmean = price.astype(int).mean()
```

```python
automobile['price'] = automobile['price'].replace('?',pmean).astype(int)
```

```python
automobile['price'].head()
```

### Cleaning the horsepower

Checking the numberic and replacing with mean value and conver the datatype to integer

```python
automobile['horsepower'].str.isnumeric().value_counts()
horsepower = automobile['horsepower'].loc[automobile['horsepower'] != '?']
hpmean = horsepower.astype(int).mean()
automobile['horsepower'] = automobile['horsepower'].replace('?',pmean).astype(int)
```

```python
automobile['horsepower'].head()
```

```python
plt.hist(automobile['horsepower'],color = 'green')
```

```python
print(automobile.loc[automobile['horsepower'] > 10000])
```

```python
automobile = automobile.loc[~(automobile['horsepower'] > 10000)]
```

```python
plt.hist(automobile['horsepower'],color = 'green')
```

### Cleaning bore



Replace the non-number value to null and convert the datatype

```python
automobile['bore'].loc[automobile['bore'] == '?']
automobile['bore'] = pd.to_numeric(automobile['bore'],errors='coerce')
```

### Cleaning the stroke

Replace the non-number value to null and convert the datatype

```python
automobile['stroke'] = pd.to_numeric(automobile['stroke'],errors='coerce')
```

### Cleaning peak-rpm

```python
automobile['peak-rpm'] = pd.to_numeric(automobile['peak-rpm'],errors='coerce')
```

### Cleaning the num-of-doors data

```python
automobile['num-of-doors'].loc[automobile['num-of-doors'] == '?']
automobile = automobile[automobile['num-of-doors'] != '?']
automobile['num-of-doors'].loc[automobile['num-of-doors'] == '?']
```

## Correlation Analysis

```python
import seaborn as sns

corr =automobile.corr()  #shows relation between different columns of a matrix
#print(corr)

plt.figure(figsize=[13,7])# creating a drawing zone
a= sns.heatmap(corr,annot=True,fmt='.2f')# this line automatically makes the heatmap, no print or out needed

#rotating the labels a little
rotx= a.set_xticklabels(a.get_xticklabels(),rotation=60)
roty= a.set_yticklabels(a.get_yticklabels(),rotation=-30)

```

```python
"""
this shows how a value in a column is related to values from other columns.
based on the data(204x24 columns of data), we provided ito the program , it has found a relation between 
them( inverly or linear ) and has given it a value. for instance the cell made by crossing city-mpg and length, the value is -0.72
this means that based on the data, the smaller the car is , the higher the milage it gives, and 72% of the data supports this?? 

"""
```

### Findings
1. Price is more correlated with engine size and curb weight of the car
2. Wheel base is highly correlated with length and width of the car
3. Curb weight is mostly correlated with engine size, length, width and wheel based which is expected as these adds up the weight of the car
