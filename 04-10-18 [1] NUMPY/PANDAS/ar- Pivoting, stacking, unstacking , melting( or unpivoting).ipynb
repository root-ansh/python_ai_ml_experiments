```python
# unpivoting/melting
"""
i think the nomenclature is wrong , because
1) it has no relation with pivoting whatsoever
2) it has no relation with multi indexing or converting multindex to singleindex ore something
3) it looks like an ugly stack, in stack we had 

    US  chevrolet chevelle malibu MPG       18
                                  year    2001
                                  xyz       v1
                                 xyz2       v2
                                 xyz3       v3
                                 ...
                                  
    US  ford torino               MPG       17
                                  year    1999
                                  xyz      v1b
                                 xyz2      v2b
                                 xyz3      v3b
                                 ..
   But here we got :
   US     ford_ranger           MPG      18.0
   US     chevy s-12            MPG      12.0
   ...
   US     ford_ranger           Year      82.0
   US     chevy s-12            Year      89.0
   
    

"""
print(table.melt(id_vars=['Origin','Model']))
table.melt(id_vars=['Origin','Model'])
```
