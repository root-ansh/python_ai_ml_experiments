```python
import sys
import pandas as pd
import numpy as np

print('python version :' + sys.version)
print('pandas version :' + pd.__version__) # its version is hidden
print('numpy version :' + np.__version__) # its version is hidden
```

```python
l1 = ('custID', [1, 21, 3, 10, 5, 67])
l2 = ('f_name', ['Ram', 'itu', 'shi', 'mii', 'tik', 'mikk'])
l3 = ('l_name', ['NICK', 'LICK', 'KICK', 'DICK', 'SICK', 'HICKK'])

l4 = ('ord_id', [100, 101, 102, 103, 104, 105])
l5 = ('custID', [1, 3, 67, 11, 32, 21])
l6 = ('amount', [100, 240, 700, 540, 620, 130])
di={
    l1[0]:l1[1],
    l2[0]:l2[1],
    l3[0]:l3[1]
}
d2={
    l4[0]:l4[1],
    l5[0]:l5[1],
    l6[0]:l6[1]
}

cframe=pd.DataFrame(di)
ord_frame=pd.DataFrame(d2)
print(cframe)
print(ord_frame)
```

```python
#basic concatintion(aka outer join or cartesian product)
# (table 1 below table 2,common columns are not repeated,
# the values of the sibling's column are filled with 'NaN')
print(pd.concat([cframe,ord_frame]))
```

```python
# 2. inner join(although it shouldn't be named that, its just an identifier for common column
# (weirly shows only the common column , i don't know why
(pd.concat([cframe, ord_frame], join="inner"))

```

```python
# 3 merge function(intersection union
# automatically merges the 2 columns where the value of custID(i.e the common column) matches
# does not show the data where column id is different

pd.merge(ord_frame, cframe)
# pd.merge(ord_frame, cframe, on='CustID') 

# #is similar . although panda can automaticlly
# identify the common column 
```

```python
pd.merge(ord_frame, cframe, how='left')
```

```python
pd.merge(ord_frame, cframe,how='right')
```

```python
pd.merge(ord_frame, cframe,how='inner')
#only the common columns
```

```python
pd.merge(ord_frame, cframe,how='outer')
#all the unique values
```
