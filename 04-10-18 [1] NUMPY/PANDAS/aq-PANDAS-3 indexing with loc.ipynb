```python
# slicing

# column position based. note they are edge case exlusive.
# i.e for rows 10:15--> 10,11,12,13,14,14 will be displayed
slice = table.iloc[4, :]
slice2 = table.iloc[20:25, 4:8]

# column name based . note they are edge case inclusive.
# i.e for rows 10:15--> 10,11,12,13,14,15 will bedisplayed 
slice3 = table.loc[10:15, ('Model', 'MPG')]
slice4 = table.loc[15:18, table.columns]
slice5 = table.loc[50:49, :]
slice6 = table.loc[50:50, :]
slice7 = table.loc[30:32, 'MPG':'Weight']
slice8 = table.loc[(20, 21, 22), 'MPG':'Weight']

print('>>>\n', slice, '\n>>>\n', slice2)
print('>>>\n', slice3)
print('>>>\n', slice4)
print('>>>\n', slice5)
print('>>>\n', slice6)
print('>>>\n', slice7)
print('>>>\n', slice8)
```

```python
# pivoting 
# showing a relation of 3 columns from the data. 
"""
eg this is the table which shows 
- all the car models ever produced in vertical
- all the manufactureres in horizontal
-the year in which manufacturer produced that such car as the  cell value
now, this can't be in some other order like year in vertical, model/make in vertical
because a manufacturer produces many cars in a year,
but a perticular created is manufactured only once. so this govis us a query like
select year,model from table where make='acura' for which it gives nan 
for all those cars the maker 'acura' didn't created and year/ modelname as:

             make=
             acura
------------+------
 CL         |2001
 integra    |1997
 series 5   |NaN           # that's made by BMW, not acura , therfore Nan
 ...
 
"""

print(table.pivot(index='model',columns='make',values='year'))
```
