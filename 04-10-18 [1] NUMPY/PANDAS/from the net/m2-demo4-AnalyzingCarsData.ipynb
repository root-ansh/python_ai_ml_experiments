```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars=pd.read_csv('../datasets/cars_data.csv')
```

```python
cars.head()
```

```python
f,ax=plt.subplots(figsize=(20,10))
sns.distplot(cars['Price'],bins=30,kde=False,rug=True);
```

```python
sns.jointplot(x='Total Salary',y='Price',data=cars,size=8)
```

```python
sns.jointplot(x='Age',y='Price',data=cars,size=8)
```

```python
sns.pairplot(cars,
            size=5,aspect=0.9,
            x_vars=['Salary','Wife Salary','Age'],
            y_vars=['Price'],
            kind='reg',
            hue='House Loan')
```

```python
sns.jointplot(x='No of Dependents',y='Price',data=cars,size=8)
```

```python
sns.pairplot(cars,size=5,aspect=0.9,
            x_vars=['Total Salary','Profession','Personal loan','Education'],
            y_vars=['Price'],
            hue='No of Dependents')
```
