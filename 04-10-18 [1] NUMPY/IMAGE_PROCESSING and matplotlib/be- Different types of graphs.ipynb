```python
# Box plot
'''
a box plot is kinda graphical representation of frequency of data
-it will majorly draw a big rectangle with a long axis coming out of its side faces.
--the long axis represents the range of most of the data, but it does not cover outliars cases(this is called whiskars).
  like if data has 9 out of 11 valuse in range 0-25, it will show an axis from 0-25 , but will show rings/dots for the remaining cases
-- the box represents the 25th and 75th percentile of data with an orange line at 50th percentage of data , or simply, the data's median.
'''
x= np.random.randint(0,25,20)
x.sort()
print(x)
plt.boxplot(x)
```

```python
x=np.append(x,28)
plt.boxplot(x)
```

```python
x=np.append(x,50)
x=np.append(x,70)
plt.boxplot(x,vert=False,notch=True)

# This will show the outliar case , horizontal box plot , and a notched pattern
```

```python
plt.boxplot(x,vert=False,notch=True,showfliers=False)
#this treats outliars as simply the data, and recalculates 25/75% and does not shows out liars
```

```python
#using patch_artist class for boxplot. it helps to access multiple properties of a boxplot
box_patch = plt.boxplot(x,patch_artist=True)
print(box_patch)

box_patch['whiskers'][1].set(color='blue',linewidth=4)

# 3 boxplots in same graph? convert dfame to a n*3 np array and see 3 box plots. more in  video :'Explore Gun Manufacturing Data Using Box And Violin Plots'
# violin plots in the same vid

```

```python
# pie charts
filename = r'C:\Users\anshs\00ANSH_PYTHON\0_DATASETS and LARGE FILES\flavours.csv'
dframe = pd.read_csv(filename)
plt.pie(dframe['Percentage'],
        labels=dframe['Flavours'],
        colors=['blue', 'gray', 'blue', 'gray', 'red'],
        autopct='%.2f',
        counterclock=False,
        explode=[0, 0.1, 0, 0.3, 0]
        )
plt.axis('equal')
```

```python
#plt . pie actually gives out a list of 3 valuse  .let's see what each of them can modify
plt.axis('equal')
wedges, texts, autotexts = plt.pie(dframe['Percentage'],
                                   labels=dframe['Flavours'],
                                   colors=['blue', 'white', 'blue', 'white', 'blue'],
                                   autopct='%.2f',
                                   counterclock=False,
                                   explode=[0, 0.1, 0, 0.3, 0]
                                   )
print('>>>', wedges,  # these are the weges objects , or the 'pastries' . we can modify each pastry individually
      '\n>>>', texts,
      '\n>>>', autotexts
      )
wedges[0].set(edgecolor='white')
wedges[1].set(edgecolor='blue')
wedges[2].set(edgecolor='white')
wedges[3].set(edgecolor='blue')
wedges[4].set(edgecolor='white')

```
