```python
"""
Matplot object hiarchy

* figure object : the main window in which matplot draws
* axis:the 2 axis x,y marks the area in which matplot will draw the diagrams/image
* "title,ticks(major and minor) and tick locations"=?
* there can be multiple graphs(marked by axis) in a figure and multiple axis in one axis
* also, everything extends from 'Artist' class. basically, its like view, viewgroup and object. everything is an object(artist) , 
  figure is kinda Viewgroup, and axis(not the x,y axis, axis means full area) is a view
* MatplotLib.pyplot is a class(or API) which offers methods/functions to work with figures and axis(i.e, higher level work)
  while rest of matplotlib offers functionality for lower levels(ticks)
  basically, Figurs-->axis-->ticks
* ANATOMY OF A FIGURE
- a figure consist of a box in which the image/graph will be drawn
- the bottommost and leftmost edges are the axis, i.e the x and y axis.
- a number of parallel lines and their intersection forms the grid in which the graph will be drawn
-the markings on the axis are called 'ticks' . eg 1,2,3,4,5 marked on y axis.
- the bold , large markings denote major ticks , which usually corresponds to gridlines. the markings between these major 
  ticks are called minor ticks.(like a scale, the 1,2,... numbers are major ticks, nd the smaller markings are minor ticks.
- Lineplots/curveplots and markers .: a graph representing a continuos data , like y=2x+3 or y=sin(x) will be lineplots/curveplots and 
  graph representing discrete data will be a marker graph. for eg , plotting (2,3),(4,3),(21,4),(3,2),... on the graph
- title : you know it.
- axis legend: basically a scale

"""
a=5
```

```python
# creating a graph with a large set of values
# np.linespace generates a np array of evenly spaced n numbers( kinda like range function)
points_count = 11  # 360
x = np.linspace(start=-5, stop=5, num=points_count)
y = np.sin(x)
# print(x, y)
plt.plot(x, y)  # its kinda shaky, but follows a pattern. increasing points_count to 360 makes it super smooth
```

```python
# we can control some ui for ticks too
plt.tick_params(axis='y', color='blue', labelcolor='orange', labelsize='15')
plt.tick_params(axis='x', bottom=False,labelbottom=False)  # bottom removes the markings, labelbottom removes the markings value

plt.plot(x, y)  # its kinda shaky, but follows a pattern. increasing points_count to 360 makes it super smooth
```

```python
# setting the legend
plt.plot(x, y, label='Sine curve')
plt.legend()
```

```python
# modifying the range of x / y or limiting the x/y
# we can zoom in/out using this function
plt.plot(x, y)
plt.xlim(-10, 10)
plt.ylim(-3.5, 3.5)
```

```python
#Multiple plots can be made using the same plot curve , but with different labels

x = np.linspace(0, 6*np.pi, 1000)
fx=np.sin(x)
gx=np.cos(x)
hx=np.tan(x)
ref=x*0

plt.plot(x, fx, label='Sine curve')
plt.plot(x,gx,label='Cos Curve')
plt.plot(x,hx,label='Tan Curve',linestyle=':')
plt.plot(x,ref,label='Refernece line')

plt.title('Various math graphs')
plt.legend()
plt.xlabel('range')
plt.ylabel('value')
plt.ylim(-2, 2)
```

```python
# annotations

fig, sub = plt.subplots()
sub.plot(
    [4, 2, 6],
    [2, 1, 5]
)
sub.annotate(
    'max val',
    xy=(6, 5),
    xytext=(5, 5.75),
    arrowprops=dict(color='m')
)
```

```python
from matplotlib import pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
A = -0.75, -0.25, 0, 0.25, 0.5, 0.75, 1.0
B = 0.73, 0.97, 1.0, 0.97, 0.88, 0.73, 0.54
plt.plot(A, B, marker='o', markerfacecolor='g', markeredgecolor='g')
for xy in zip(A, B):
    ax.annotate('(%s, %s)' % xy, xy=xy, textcoords='data')
plt.grid()
plt.show()
```

```python
# scaling the x/y axis
np.random.seed()
y=np.random.sample(size=5000)
y.sort() # this makes y an array of random numbers in increasing order


x=np.arange(5000)
plt.grid(True)
plt.plot(x,y)


```

```python
# scaling the x/y axis
np.random.seed()
y=np.random.sample(size=5000)
y.sort() # this makes y an array of random numbers in increasing order
plt.yscale('log', basey=2) # we changed the y scale to the powers of log2
x=np.arange(5000)
plt.grid(True)
plt.plot(x,y)

# same data, different representation
```

```python
# scaling the x/y axis
np.random.seed()
y=np.random.sample(size=5000)
y.sort() # this makes y an array of random numbers in increasing order
plt.xscale('log', basex=8) # we changed the x scale to the powers of 2
x=np.arange(5000)
plt.grid(True)
plt.plot(x,y)

# same data, different representation
```

```python
from matplotlib.ticker import NullFormatter

plt.grid(True)

vals = np.random.uniform(low=0, high=1, size=1200)
vals.sort()

plt.yscale('logit')  # changes y scale into log values i guess?
plt.gca().yaxis.set_minor_formatter(
    NullFormatter())  # we access the yaxis out of the set of 2 axis and removes the label of all  axis. thus we have lesser log values
                      # gca() = get_current_axis --> (axis1, axis2,...)
                      
plt.plot(np.arange(len(vals)), vals)
```

```python
# Using multiple axis for same data: twin axis
filename= r'C:\Users\anshs\00ANSH_PYTHON\0_DATASETS and LARGE FILES\weather_madrid_LEMD_1997_2015.csv'
dframe= pd.read_csv(filename)

fil_ter=dframe[['CET', 'Mean TemperatureC', 'Mean Wind SpeedKm/h']].head(100)
fil_ter['CET']=pd.to_datetime(fil_ter['CET'])
fil_ter
```

```python
fig, sub_tempC = plt.subplots()
fig.set_figwidth(10)
fig.set_figheight(5)

sub_tempC.grid()
sub_tempC.set_xlabel('Date')
sub_tempC.tick_params(axis='x', bottom=False, labelbottom=False)

sub_tempC.set_ylabel('Temp(c)', color='red', size='x-large')
sub_tempC.tick_params(axis='y', labelcolor='red', labelsize='large')

sub_tempC.plot(fil_ter['CET'], fil_ter['Mean TemperatureC'], color='r')

sub_new = sub_tempC.twinx()
sub_new.set_ylabel('AVERAGE WIND SPEED(KMPH)', color='g', size='x-large')
sub_new.tick_params(axis='y', labelcolor='g', labelsize='large')

sub_tempC.plot(fil_ter['CET'], fil_ter['Mean Wind SpeedKm/h'], color='g')

```
