```python
# patches module is used to draw various shapes , irrespective of the data.

#1 drawing a rectangle
fig, sub = plt.subplots()
sub.add_patch(patches.Rectangle((0.4, 0.2), 0.5, 0.5, fill=False))
sub.add_patch(patches.Rectangle((0.1, 0.1), 0.4, 0.4, fill=True))
sub.add_patch(patches.Rectangle((0.3, 0.3), 0.5, 0.5, 30.0, color='m'))  # tis is a square but looks like rhombus
sub.add_patch(patches.Rectangle((0.2, 0.2), 0.2, 0.2, facecolor='black', edgecolor='yellow'))  # tis is a square but looks like rectangle
```

```python
#1 more rectangles, in a better grid
fig, sub = plt.subplots()
sub.set_aspect('equal')

sub.add_patch(patches.Rectangle((0.3, 0.3), 0.5, 0.5, 30.0, color='m'))# tis is a square 
sub.add_patch(patches.Rectangle((0.2, 0.2), 0.2, 0.2,color='black'))# tis is a square 
sub.add_patch(patches.Rectangle((0.5,0.2),0.3,0.6,hatch='x',fill=False))# a designer rectangle
sub.add_patch(patches.Rectangle((0.1,0.2),0.3,0.6,hatch='*'))# a designer rectangle
# some other hatches :  | , -



```

```python
# making a cool graph

fig, sub = plt.subplots()
sub.set_aspect('equal')
# ticks = np.arange(0, 1.6, 0.1)
# plt.xticks(ticks)
# plt.yticks(ticks)

sub.set_xlim(0, 1.6, 0.1)
b1 = patches.Rectangle([0.1, 0.2], 0.2, 0.6, alpha=0.2)
b2 = patches.Rectangle([0.5, 0.2], 0.2, 0.6, alpha=0.5)
b3 = patches.Rectangle([0.9, 0.2], 0.2, 0.6, alpha=1)
b4 = patches.Rectangle([1.3, 0.2], 0.2, 0.6, alpha=None)

patch_list = b1, b2, b3, b4
for p in patch_list:
    sub.add_patch(p)

# more attributes facecolor, edgecolor,fill,edgewidh(range:0-10), linestyle(dashed,solid,dashdot,dotted),hatch
```

```python
# circles

fig, sub = plt.subplots()
sub.set_aspect('equal')
# ticks = np.arange(0, 1.6, 0.1)
# plt.xticks(ticks)
# plt.yticks(ticks)

sub.set_xlim(0, 1.6, 0.1)
b1 = patches.Circle([0.3, 0.4], 0.1, hatch='*')  # radius of 0.2
b2 = patches.Circle([0.5, 0.2], 0.2, hatch='-', facecolor='red')
b3 = patches.Circle([0.7, 0.4], 0.2, hatch='x', facecolor='plum')
b4 = patches.Circle([0.5, 0.7], 0.2, hatch='+', fill=False)

patch_list = b1, b2, b3, b4
for p in patch_list:
    sub.add_patch(p)
```

```python
# making a polygon
fig, sub = plt.subplots()
sub.set_aspect('equal')

patch_poly = patches.Polygon(
    [
        [0.8, 0.8], [0.7, 0.1],
        [0.3, 0.1], [0.2, 0.8],
        [0.5, 0.5]
    ]
)

sub.add_patch(patch_poly)
```

```python
# making a polygon
fig, sub = plt.subplots()
sub.set_aspect('equal')

patch_poly = patches.Polygon(
    [
        [0.8, 0.8], [0.7, 0.1],
        [0.3, 0.1], [0.2, 0.8],
        [0.5, 0.5]
    ],
    fill=False,closed=False
)

sub.add_patch(patch_poly)
```

```python
# bezier curve : a mathemetical way to draw any curve existing. these curves are determined by  nodes or handles which
#  can make their curve depth change whenever we want
'''
         A
       /  O2
      / /  \
     / o    \
    //       \
   O1         \
  /            \
 B             C
O1 goes from B to A
O2 goes from A to C
o is responsible for making curve, as it goes from O1 to O2 while O1 and O2 are themselves moving
'''
pi=3.14
```

```python
from matplotlib import pyplot as plt
from matplotlib import patches as patches
from matplotlib.path import Path # imp, we are importing 'Path' Function of 'path' module in matplotlib. python's naming is pretty fucked up :|
fig, sub = plt.subplots()

p = patches.PathPatch(
    Path(
        [
            (0.1, 0.1), (0.2, 0.8),
            (0.8, 0.1), (0.8, 0.7)
        ],
        [
            Path.MOVETO, Path.LINETO,
            Path.LINETO, Path.LINETO
        ]
    ),
    fill=None
)
sub.add_patch(p)
```
