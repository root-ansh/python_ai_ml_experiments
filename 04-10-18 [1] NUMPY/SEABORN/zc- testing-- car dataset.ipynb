```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print(sns.__version__)
print(np.__version__)
print(pd.__version__)

filename=r'C:\Users\anshs\00ANSH_PYTHON\0_DATASETS and LARGE FILES\cars_data.csv'
dframe =pd.read_csv(filename)
dframe.head()
```

```python
sns.distplot(dframe['Price'], bins=30, kde=False, rug=True)
```

```python
# Q How does the consumer buying behavior vary by total income of the household?
sns.jointplot(x='Total Salary', y='Price', data=dframe)
```

```python
# does age of the buying person has anything to do with price of car?
sns.jointplot(x='Age', y='Price', data=dframe)
```

```python
sns.pairplot(
    dframe,
    height=5, aspect=0.9,
    x_vars=['Salary', 'Wife Salary', 'Age'],
    y_vars=['Price'],
    kind='reg',
    hue='House Loan'
)
sns.pairplot(
    dframe,
    height=5, aspect=0.9,
    x_vars=['Salary', 'Wife Salary', 'Age'],
    y_vars=['Price'],
    kind='reg',
    hue='House Loan'
)
```

```python
sns.jointplot(x='No of Dependents', y='Price', data=dframe)
```

```python
sns.pairplot(
    dframe,
    height=5, aspect=0.9,
    x_vars=['Total Salary', 'Profession', 'Personal loan', 'Education'],
    y_vars=['Price'],
    hue='No of Dependents'
)
```

```python
# we can set the look nd feel of seaborn in 5 different styles:
# sns.set_style('whitegrid') # most beautiful
# sns.set_style('darkgrid')  # system default
# sns.set_style('white')       # jupyter default
# sns.set_style('dark')
sns.set_style('ticks')          # similar to white style, but with prominant ticks
f, sub = plt.subplots(figsize=[10, 5])
sns.distplot(dframe['Price'])
```

```python
# to use a style for only a single graph , and not for whole grid, we use with keyword along with axes_style functions
with sns.axes_style('dark'):
    f, sub = plt.subplots(figsize=[10, 5])
    sns.distplot(dframe['Price'])
```
