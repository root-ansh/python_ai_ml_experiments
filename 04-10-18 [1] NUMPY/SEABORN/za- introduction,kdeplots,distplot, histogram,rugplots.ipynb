```python
# plotting a frequency distribution histogram or or simply distribution plot
sns.distplot(dframe['Calories'])
```

```python
# ^^^^^^^
# we can se tha mostly calories lies beterr 250-350, and their are outliars around 1800
# there is also a line graph, that's kde plot
```

```python
# removing the kde plot, we get a more clearer histogram
# increasing the bin size , reduces the  y values , because each bin now shows more precise data
# for eg , initally bin 0-100 showed ~50 values, but now we have a bin at 0-30 which shows 17 values
#if we used 260 bins, i.e the length of data , we migh have only 0 and 1 on y(wanna try?)

sns.distplot(dframe['Calories'],kde=False,bins=50)
# sns.distplot(dframe['Calories'],kde=False,bins=2)
# sns.distplot(dframe['Calories'],kde=False,bins=260)
```

```python
#rug plot, a plot showing a line for  each value in dataset
sns.rugplot(dframe['Calories'],axis='x',height=0.7)
```

```python
#best usecase is to use with distribution plot
sns.rugplot(dframe['Calories'],height=0.6,axis='x')
sns.distplot(dframe['Calories'], kde=False )
```

```python
# distplot also have a built in rug attribute , so no need for specific rugplot calling
# ( although , when called speecifically, we have better customisation options)

sns.distplot(dframe['Calories'], kde=False, rug=True, bins=20)
```

```python
# checking on with kde plot:

sns.distplot(dframe['Calories'], rug=True, kde=True,hist=False)
```

```python
# customising sns env
# we are adding a grid using sns.set(color_codes=True) and then calling 
# kdeplot specifically using kdeplot(...), coz it will give us more customization optiobns, like shaded graph and colors
# notice that we don't have to call legend to show legend
sns.set(color_codes=True)
sns.kdeplot(dframe['Calories'], shade=True,color='yellow')
```

```python
# modifieng the bandwidth of kde plot
plt.figure(figsize=[15,5])
sns.kdeplot(dframe['Calories'], color='red', bw=10,label='bw:10')
sns.kdeplot(dframe['Calories'], color='blue', bw=30,label='bw:30')
sns.kdeplot(dframe['Calories'], color='green', bw=150,label='bw:150')

sns.kdeplot(dframe['Calories'], color='orange')

# as we see , with decreasing bandwidhth, we get a more accurate kde curve, but the general trend is harder to see
```

```python
# with 0 band width its just a  graph showing all the actual calaroie values
sns.kdeplot(dframe['Calories'], color='orange',bw=00)
```
