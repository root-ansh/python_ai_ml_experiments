```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print(sns.__version__)
print(np.__version__)
print(pd.__version__)
```

```python
filename=r'C:\Users\anshs\00ANSH_PYTHON\0_DATASETS and LARGE FILES\menu.csv'
dframe =pd.read_csv(filename)
print(dframe.head())
```

```python
#L-M plot
"""
this is one of  the basic kinds of machine learning. here relation between 2 cols x,y is plotted as a scatter graph
a line passes through them which represents the expected next values based on the current data.
the shade alongside  represents the 'confidence area' i.e, the value won't lie outside this area , acc to the curent pattern
it is lesser in the area with the dots , i.e  algorithm doesn't expect the values to lie out side the scatter plot's area , but
it increases as the range increases to the unknown area, because the algorithm doesn't have a confidence in the line it has plotted.
i don't know what each  attribute means , but let's try changing them ;)
"""
sns.lmplot(x='Calories', y='Total Fat', data=dframe, height=7, aspect=2)
# sns.lmplot(x='Calories', y='Total Fat', data=dframe,height=17, aspect=2) # weird blurred graph
# sns.lmplot(x='Calories',y='Total Fat',data=dframe,height=7,aspect=3)# weird blurred graph
```

```python
#(changing x,y axis and adding a jitter)
sns.lmplot(y='Calories', x='Total Fat', data=dframe, height=7,x_jitter=.1)
```

```python
# adding an x_estimator
sns.lmplot(y='Calories', x='Total Fat', data=dframe,x_estimator=np.mean)
```

```python
# addig a column value
sns.lmplot( x='Total Fat',y='Calories', data=dframe,col='Trans Fat')
```

```python
#plotting the data as 3 dimensional with 'Category' column as hue
sns.lmplot(x='Total Fat', y='Calories', data=dframe, col='Trans Fat', hue='Category')
```

```python
# viewing the data in a row
sns.lmplot(x='Total Fat', y='Calories', data=dframe, row='Trans Fat', hue='Category')
```

```python
# customising seaborne grids
sns.set(style='whitegrid', color_codes=True) # setting a whitegrid
```

```python
# catagorical and Multipanel data visualisation

# 1 strip plot
'''
    i don't know, but its kinda histogramic plot, it plots the relation between categories and calories  by catagorizing them in bins made of 
    category values.

'''

np.random.seed(sum(map(ord, 'catagorical')))
f, sub = plt.subplots(figsize=[15, 5])
sns.stripplot(x='Category', y='Calories', data=dframe, size=6)

```

```python
# using jitter=True makes it:
sns.stripplot(x='Category', y='Calories', data=dframe, size=6,jitter=True)
```

```python
# swarmplot is similar to strip pot, only it does not allow overlapping of content, by placing it horizontally
# although i feel this as kinda wrong, because if we  consider the space between , say 1th and 2th col
# as 1.1,1.2,1.3... , it plots 2 [1,5]points as [0.9,5] and [1,5]
# but meh.
fig = plt.subplots(figsize=[15, 6])
sns.swarmplot(x='Category', y='Calories', data=dframe)
```

```python
# adding a third column or 'hue', t plots hue vs catagory too .
# now the position of dot represent relation between category and calorie
# and the color shows transfat
sns.swarmplot(x='Category', y='Calories', data=dframe, hue='Trans Fat')
```

```python
# changing x,y, we can see a better distributuion?meh.
sns.swarmplot(y='Category', x='Calories', data=dframe, hue='Trans Fat')
```

```python
# facet grids: for plotting multi dimensional relations
# what i understood: from the looks of it, it seems that Facet grid is kinda grid maker for all values of a perticular column
# for eg, House Loan had only 2 values yes and no, so it made a 2 grids.
# the object returned by this function has javascript-like methods which can map fuction definations.
# here , the histogram of houseloan's yes is mapped with price data
# cool
g = sns.FacetGrid(dframe, col='House Loan')
g.map(plt.hist, 'Price')
```

```python
# customizing facetgrids

g = sns.FacetGrid(dframe, col='No of Dependents', col_wrap=2, aspect=3)
g.map(sns.barplot, 'Price', 'Total Salary', color='#D7BDE2', edgecolor='red', lw=.5)
g.fig.subplots_adjust(wspace=1, hspace=0.5)
g.set_axis_labels('PriceOfCARS', 'HH INCOME')
g.set(yticks=[500000, 1000000, 1500000, 2000000])
```

```python
filename=r'C:\Users\anshs\00ANSH_PYTHON\0_DATASETS and LARGE FILES\cars_data.csv'
dframe =pd.read_csv(filename)
dframe.head()
```

```python
# pairgrids: kinda similar to facet grids

g = sns.PairGrid(dframe, vars=['Price', 'Salary', 'Wife Salary'])
g.map(plt.scatter)

# univariate relaion ships : when y==x eg prics vs price graph . bivariate: when x!=y
```

```python
# this is kinda tricky to explain.
g = sns.PairGrid(dframe, vars=['Price', 'Salary', 'Wife Salary'])
g.map_offdiag(plt.scatter)
g.map_diag(sns.kdeplot)
g.add_legend()
```

```python
# this is again kinda tricky to explain.
g = sns.PairGrid(dframe, vars=['Price', 'Salary', 'Wife Salary'],hue='House Loan')
g.map_offdiag(plt.scatter)
g.map_diag(sns.kdeplot)
g.add_legend()
```

```python
# this is again kinda tricky to explain.
g = sns.PairGrid(dframe, vars=['Price', 'Salary', 'Wife Salary'],hue='House Loan')
g.map_diag(sns.kdeplot)
g.map_upper(plt.scatter)
g.map_lower(sns.regplot)
g.add_legend()
```

```python
g = sns.PairGrid(dframe, x_vars=['Price', 'Total Salary'], y_vars=['Salary', 'Wife Salary'], hue='House Loan')
g.map(plt.scatter)
g.add_legend()
```
