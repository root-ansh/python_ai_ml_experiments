```python
# count all NaN's for a column
# print(
#     '>>>\n',type(dframe2['FuelEconomy'].isnull()),'\n',dframe2['FuelEconomy'].isnull()
# )
print(len(dframe2))
print(
    '>>>\n', type(dframe2['FuelEconomy'].isnull().sum()), dframe2['FuelEconomy'].isnull().sum()
)
print(
    '>>>\n', type(dframe2['FuelEconomy'].isnull().sum()), dframe2['FuelEconomy'].notnull().sum()
)

# output1 : boolean series for fueleconomy. i.e a series of True/false values for said filter
# output2 : counts of all the true vals
# output3 : counts of all the false vals (not null function )
```

```python
#1 replacing NaN's Everywhere with 0's ('note : this function sees ALLL the NaN's AND NOT JUST ONE COLUMN)
print(dframe2.fillna(0))
```

```python
#2 dropping the complete rows if data is nan
print(len(dframe2.dropna()))
print(dframe2.dropna())
```

```python
# checking if table has any nan's in any column

print(dframe2.isna().sum())
print('='*20)
print(dframe2.dropna().isna().sum())
```

```python
# replacing values in complete table
# will work for complete words , not patterns (eg will change 'root' to 'chi' if 'root' is solo entry , but wont replace 'root' in 'rootansh'
# or 'this is a root of tree' 

#print(dframe2.replace({'Monday': 'MON', 'Friday': np.nan}))
print(dframe2.replace({'Monday': 'MON', 'Friday': np.nan}).isnull().sum())
```
