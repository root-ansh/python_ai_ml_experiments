## series.

'''
A `Series` is a 1-d representation of a list or a numpy array.  
we can pass a list or np-array to it and it will store the data as **index-value pairs** , where both the index as well as  the values are modifieable.


'''



````python
data="J F M A M J J A S O N D".split(' ')
print(">>>",data)

series= pd.Series(data)
#print(series)

n_arr= np.array(data)
print(pd.Series(n_arr))


````

````python
column_labels="id,name,age,gender".split(',')
row_1=11,'ram',19,'M'

sales_db=pd.Series(data=row_1,index=column_labels)
print(sales_db)

````

````python
# panda series can also be initalised using a dictionary

di = {
    'id': [22, 23, 24, 25],
    'name': list('abcd'),
    'age': [19, 20, 20, 19],
    'gender': list("MFMM")
}
db = pd.Series(di)
print('>>>', db, "type=", type(db))
print(">>>", db.index)
print(">>>", db.values)

print(">>>", db[['name', 'age']])
````

## dataframe.
'''
A `Dataframe` is another representation of a list or a numpy array.  
we can pass a list or np-array to it and it will store the data as **2d-matrix** where **data** , **vertical_index** and **Horizontal_index**,are all modifieable.  

```
+----------+----------+----------+----------+
|          |--col_i1--+--col_i2--+--col_i3--+
+----------+----------+----------+----------+
|--idx_i1--|   value  |   value  |   value  |
+----------+----------+----------+----------+
|--idx_i2--|   value  |   value  |   value  |
+----------+----------+----------+----------+
|--idx_i3--|   value  |   value  |   value  |
+----------+----------+----------+----------+

```


'''

````python
# basics 

dframe=pd.DataFrame()
print(">>>",type(dframe),dframe)

values=list("abcdef")
dframe2=pd.DataFrame(values)
dframe2
````

````python
names="abc bcd cde def".split(' ')
values=2,1,3,7
data= list(zip(names,values))
print(data)
dframe3=pd.DataFrame(data,columns=('name','val'))
print(dframe3)#it looks ugly when used with print
dframe3
````

````python
#mistake data 
names="abc bcd cde def".split(' ')
values=['idx1','idx2','idx3','idx4']
dframe3=pd.DataFrame(names,values)
dframe3
````

````python
idx_names=['idx1','idx2','idx3','idx4']
col_names=['col1','col2','col3','col4']
data=np.random.randint(5,10,16).reshape(4,4)
dframe4=pd.DataFrame(data,index=idx_names,columns=col_names)
print(dframe4)
dframe4

````

````python
# dataframe can take a dict as input and it willl consider it as 
# a set of columns where col name= key ,columns data= val and it will give default value(0,1,..)
# as indexes
di = {
    'search': ("a1,a2,a3,a4".split(',')),
    'sort': ("b1,b2,b3,b4".split(',')),
    'ML': ("v1,v2,v3,v4".split(',')),
    'AI': ("x1,x2,x3,x4".split(',')),
}
dframe =pd.DataFrame(di)
dframe

````

````python
# what happens when we use it with panda series?nothing,
di = {
    'search': pd.Series("a1,a2,a3,a4".split(',')),
    'sort': pd.Series("b1,b2,b3,b4".split(',')),
    'ML': pd.Series("v1,v2,v3,v4".split(',')),
    'AI': pd.Series("x1,x2,x3,x4".split(',')),
}
dframe =pd.DataFrame(di)
dframe
````
