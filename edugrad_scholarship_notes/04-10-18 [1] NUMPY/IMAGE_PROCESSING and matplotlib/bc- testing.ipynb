```python
from matplotlib import pyplot as plt
import numpy as np
import sys
import pandas as pd

print('versions:')
print('system:', sys.version)
print('numpy:', np.version)
print('pandas:',pd.__version__)
# print('pypl0t:',plt.__version__)
```

```python
filename = r'C:\Users\anshs\00ANSH_PYTHON\0_DATASETS and LARGE FILES\AAPL-2017.csv'

dframe = pd.read_csv(filename)
dframe.head()
```

```python
dframe['my_dates']=pd.to_datetime(dframe['Date'])
dframe.head()
```

```python
fig = plt.figure(figsize=[10, 6])
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(dframe['Date'],dframe['Open'],color='r')
ax1.set_title('Open vs close(inset)')

ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])
ax2.plot(dframe['Date'],dframe['Close'],color='purple')

```

```python
fig = plt.figure(figsize=[10, 6])
fig.suptitle('Stock Price 2017',fontsize=25)

sup1=fig.add_subplot(221)
sup1.set_title('OPEN')
sup1.plot(dframe['Date'],dframe['Open'])

sup1=fig.add_subplot(222)
sup1.set_title('CLOSE')
sup1.plot(dframe['Date'],dframe['Close'])


sup1=fig.add_subplot(223)
sup1.set_title('Highest')
sup1.plot(dframe['Date'],dframe['High'])


sup1=fig.add_subplot(224)
sup1.set_title('Volume')
sup1.plot(dframe['Date'],dframe['Volume'])

```
