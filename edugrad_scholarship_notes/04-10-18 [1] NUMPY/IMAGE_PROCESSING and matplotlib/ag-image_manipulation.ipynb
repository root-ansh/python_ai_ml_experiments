```python
# in a single dimension , the rgb representation can be different than simple [r,g,b]= [(0:255),(0:255),(0:255)]
# as we saw in last example , the random array generated b/w 0,1 also have colors to images
# because  scikit has this kind of notation written in it.
#   "if your data is in floating point.it must be between [0,1] {both inlueded),
#   if data is out of that bounds , it must be integer between [0,255]."
"""
Thus we can have different types of n-dimensional arrays which have different meanings.
2d grayscale    --  [[141,222,12,3,1],[0,0,4,3,2],...]  --  a 2d  array representing intensity of black color as [[pixel1,p2,p3][p4,p5,p6],...]
                                                            where p1,p2,p3... are different pixels on the screen. their data type can be:
                                                                    --integer : b/w 0 & 255 : 0>>pure black, 255>>pure white .eg [147,3,0,254,..]
                                                                    --float   : b/w 0 & 1  : 0>>pureblack, 1>>pure white. eg [0.144255,0.98987,..]
2d colored      -- [ [ [21,3,4],[33,254,1],... ] ,      -- a 3d array representing a color in rgb value such that [ [ [--pix1--][pix2],..]
                     [ [21,3,4],[32,254,1],...]                                                                        ,...],[[..]],...]
                     ,...                                  such that the smallest column represents the color intensity of a single pixel
                    ]
3d greayscales
3d colored
"""
```

```python
linear0 = np.linspace(0, 1, 25).reshape((5, 5))
linear1 = np.linspace(0, 100, 100).reshape((10, 10)).astype(np.uint8)

print("Linear0:", type(linear0), linear0.dtype, linear0.min(), linear0.max())
print("Linear1:", type(linear1), linear1.dtype, linear1.min(), linear1.max())
print(">>>\n", linear0)
print(">>>\n", linear1)

fig, (ax0, ax1) = plt.subplots(1, 2)
ax0.imshow(linear0, cmap='gray')
ax1.imshow(linear1, cmap='gray');

"""
multiple things are happenning here .
1) linespace(start,stop,total) is a function to generate an linear numpy-array having equal differences(in an Arithmetic-progression).
2) .astype(np.uint8) ensures that the numbbers generated by the linespace(..) are all  rounded off to 8-bit integral values.
3)???
-->fig, (ax0, ax1) = plt.subplots(1, 2)
-->ax0.imshow(linear0, cmap='gray')
-->ax1.imshow(linear1, cmap='gray');
(#see in[138])

PS:
the original values for linear1 were linear 1 = np.linspace(0, 255, 25).reshape((5, 5)).astype(np.uint8). this was taken to show that for a 
given size , linear interpolation(i.e, adding a no. of elements between 2 numbers such that the difference between all remains constant or simply
AP) for 0 and 1 is same as that of 0 and 255.
i changed it for fun.

"""
```

```python
#More basic image operations:

#---->cropping and merging

cat = data.chelsea()
print("stats cat:", type(cat), cat.dtype, cat.min(), cat.max(), cat.shape, cat.size)

#cropping
cat_crop1 = cat[100:, 50:, :]  # selects all rows above 100th row , columns above 50th
plt.imshow(cat_crop1)
#horizontal splitting
cat_crop2, cat_crop3 = np.split(cat, 2)  #splits the picture row wise(i.e in 2 horizontal rctangles                 s
plt.imshow(cat_crop2)
print("stats cat2:", cat_crop2.dtype, cat_crop2.min(), cat_crop2.max(), cat_crop2.shape)
plt.imshow(cat_crop3)
print("stats cat3:", cat_crop3.dtype, cat_crop3.min(), cat_crop3.max(), cat_crop3.shape)

# vertical splitting
#cat_crop2,cat_crop3=np.split(cat,2,axis=1)#splits the picture column wise(i.e in 2 vertical rctangles                 s
# but this will give error because total columns=451 which is not splittable into 2 equal vertical rectangles                                      
# therefore,
cat = cat[:, :450, :]# cropped one pixel width screen
cat_crop2, cat_crop3 = np.split(cat, 2, axis=1)  # now this will work correctly
plt.imshow(cat_crop2)
print("stats cat2:", cat_crop2.dtype, cat_crop2.min(), cat_crop2.max(), cat_crop2.shape)
plt.imshow(cat_crop3)
print("stats cat3:", cat_crop3.dtype, cat_crop3.min(), cat_crop3.max(), cat_crop3.shape)

#horizontal merging: place the two pics one upon the other
plt.imshow(np.concatenate((cat_crop2,cat_crop3)))

#vertical merging: place the two pics side by side
plt.imshow(np.concatenate((cat_crop2,cat_crop3),axis=1))


#---->masking
cameraman = data.camera()
#plt.imshow(cameraman)
# actual picture is b/w , but being actively masked by cmap with
#  blue-yellow vigriti color. actual picture is blue yellow      
print("stats:", cameraman.dtype, cameraman.min(), cameraman.max(), cameraman.shape)
mask = cameraman < 87
cameraman[mask] = 255

#plt.imshow(mask)  # mask = yellow + putple bg
#plt.imshow(cameraman)  # masked cameraman = yelllow with vigriti blue background

#---->Fancy indexing (making fancy design on image):

camera = data.camera()
inds_r = np.arange(len(camera))
inds_c = 10 * inds_r % len(camera)
camera[inds_r, inds_c] = 255
print("stats:", camera.dtype, camera.min(), camera.max(), camera.shape)
#plt.imshow(camera, cmap='gray')
```
