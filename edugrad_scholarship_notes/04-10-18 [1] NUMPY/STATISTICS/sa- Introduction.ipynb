1. mean = average of all the data , i,e : 
           m = (x1+x2+....xn)/n  , where m is mean and n is total elements
2. range=  xmax - xmin  
3. mean deviation(D) = [ (x1-m) + (x2-m) + (x3-m)..... ]
4. square mean deviation(Dsq) = [ (x1-m)^2 + (x2-m)^2 + (x3-m)^2..... ]
5. Variance = [ (x1-m)^2 + (x2-m)^2 + (x3-m)^2..... ]/n  == Dsq/n
6. Variance with bassels correction = [ (x1-m)^2 + (x2-m)^2 + (x3-m)^2..... ]/n-1  == Dsq/n-1
7. std. deviation = (Variance)^0.5

```python
# Comparing the t-statistics

#GENERALISING THE SAMPLES WITH HYPOTHESIS TESTING
# let x be a 'sample' i.e , x is a small list of data taken from a larger list of data X, to represent the bigger list X.
# representative sample : a sample which can really represent the bigger data
# biased sample : a sample which does not represent the original data , and is rather biased towards a perticular attribute.
# eg : in a set of 100 people with 70 males and 30 females, 4 people are choosen as sample. the 4 people are 2 boys and 2 girls which shows that
# 50% ppl are boys and 50% pppl are girls , which is clearly wrong, since  70% ppl are boys and 30% ppl are girls.

# cookie cutter hypothesis testing
#
```

```python
np.mean(dframe.age)
stats.ttest_1samp(dframe.age, 38.6435)  # Ttest_1sampResult(statistic=0.0013772033081731628, pvalue=0.9989011567153004) 
# ^^^ gives a very small 'statistic' value and a high 'pvalue' . therefore the chances of the next number to be 38.6435 value are pretty high


stats.ttest_1samp(dframe.age, 48.6435)  # Ttest_1sampResult(statistic=-161.19048472493145, pvalue=0.0)
# ^^ gives a -ve stats value and 0 as pvalue, meaning the 48.6435 is above the mean

stats.ttest_1samp(dframe.age, 8.6435)  # Ttest_1sampResult(statistic=483.5769629880271, pvalue=0.0)
# ^^ gives a -ve stats value and 0 as pvalue, meaning the 8.6435 is below the mean


y1 = [1, 2, 3, 4, 5]
y2 = [1, 2, 3, 4, 5, 6]
stats.ttest_ind(y1, y2)  # Ttest_indResult(statistic=-0.47237749297333015, pvalue=0.6479015995584663)
# shows there is some kind of similarity


y1 = [1, 2, 3, 4, 5, 6]
y2 = [1, 2, 3, 4, 5, 6]
stats.ttest_ind(y1, y2)  #Ttest_indResult(statistic=0.0, pvalue=1.0)

y1 = [1, 2, 3, 4, 5, 6]
y2 = [6, 5, 4, 1, 2, 3] # order of numbers doesnt matter , as long as they are all equal
stats.ttest_ind(y1, y2)  #Ttest_indResult(statistic=0.0, pvalue=1.0)

# ----------------- ttest_rel


y1 = [1, 2, 3, 4, 5, 6]
y2 = [6, 5, 4, 1, 2, 3] # order of numbers doesnt matter , as long as they are all equal
stats.ttest_rel(y1, y2)  #Ttest_relResult(statistic=0.0, pvalue=1.0)

#but
y1 = [1, 2, 3, 4, 5, 6]
y2 = [1, 2, 3, 4, 5, 6]
stats.ttest_rel(y1, y2)  # Ttest_relResult(statistic=nan, pvalue=nan)
```

```python
y1 = [1, 2, 3, 4, 5, 6]
y2 = [1, 1, 1, 1, 1, 1]
y3 = [1, 2, 1, 3, 1, 1]
y4 = [12, 13, 14, 15, 16, 17]

print(stats.f_oneway(y1, y2, y3))  # F_onewayResult(statistic=7.500000000000001, pvalue=0.005524271728019903)
print(stats.f_oneway(y1, y2, y4))  # F_onewayResult(statistic=132.6428571428567, pvalue=2.908554595423212e-10)
```
