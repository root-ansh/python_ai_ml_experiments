```python
#Q1

# basic linked list with only  append function, length function and string reprsentation function    
class node:
    def __init__(self,data=None,nextnode=None):
        self.data=data
        self.nextnode=nextnode
    def __repr__(self):
        return repr(self.data)

class LinkedList:     
    def __init__(self):
        self.head = None

    def appendData(self, data):

        if not self.head:
            self.head = node(data=data)
            return
        curr = self.head
        while curr.nextnode:
            curr = curr.nextnode
        curr.nextnode = node(data=data)
               

    def length(self):
        if (self.head  == None): 
            return  0
        else: 
            curr = self.head
            count = 0
            while (curr != None):
                curr = curr.nextnode
                count += 1            
        return count

    
    def __repr__(self):
        nodes = []
        curr = self.head
        while curr:
            nodes.append(repr(curr))
            curr = curr.nextnode
        return '[' + ', '.join(nodes) + ']'


#Driver program:
l=LinkedList()
print("initially , the list is :\t",l)
print("size :\t",l.length())

l.appendData(50)
l.appendData("HAHAHA")
l.appendData(True)
print("after appending , the list is :\t",l)
print("size :\t",l.length())


```

```python
#Q3 A basic queue with functions : enque, deque, size , and isempty() function . since this will use a list which doesn't have a fixed size in python , therefore it would never give an overflow exception

class Myqueue:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def enq(self, item):
        self.items.append(item)

    def deq(self):
        if(len(self.items)>=1):
            return self.items.pop(0)
        else:
            pass# deq won't work when  queue is empty

    def size(self):
        return len(self.items)
    def __repr__(self):
        return(str(self.items))

    

    
#driver program
q=Myqueue()
print(q)
print("size:",q.size())

q.enq(0)
print(q)

q.enq("second")
print(q)

q.enq({"third":3})
print(q)

q.enq(True)
print(q)

print()
print("removed:",q.deq(),"\t\t\tqueue is now:",q)
print("removed:",q.deq(),"\t\tqueue is now:",q)
print("removed:",q.deq(),"\t\tqueue is now:",q)
print("removed:",q.deq(),"\t\t\tqueue is now:",q)



print(q)




```

```python
#Q2 A stack class using  Queue.
"""
This stack uses Myqueue from the solution of Q3.
it has the functions push ,pop, peek, top, and display(using __repr__ function)
"""

class stack:
    def __init__(self):
        self.dataholder=Myqueue()
        
    def push(self, val):
        self.dataholder.enq(val)
    def pop(self):
        tmpQ= Myqueue()
        for i in range(self.dataholder.size()-1): # iterates over all elements except the last one
            tmpQ.enq(self.dataholder.deq()) 
            """
            since queue uses fifo, it won't matter if we pass the element
            removed from main queue to temporary queue, we will still have the same queue
            (except the last item, which is still in the main queue)

            """
        valtobepopped=self.dataholder.deq() # removing the last value from the queue
        self.dataholder = tmpQ
        return valtobepopped

    def isEmpty(self):
        return self.dataholder.size == 0
    def size(self):
        return self.dataholder.size()
        
        
    def __repr__(self):
        return str(self.dataholder)
        
    
#driver program
s=stack()
print(s)
print("size:",s.size())

s.push(0)
print(s)

s.push("second")
print(s)

s.push({"third":3})
print(s)

s.push(True)
print(s)

print()
print("removed:",s.pop(),"\t\t\tstack is now:",s)
print("removed:",s.pop(),"\t\tstack is now:",s)
print("removed:",s.pop(),"\t\tstack is now:",s)
print("removed:",s.pop(),"\t\t\tstack is now:",s)


print(q)

```

```python
#Q4 sorting using bubble sort

def bubbleSort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if (arr[j] > arr[j+1]) :
                arr[j], arr[j+1] = arr[j+1], arr[j]
                
                
list1 = [80,70,90,40,60,50]

print("list1 before sorting :",list1)
bubbleSort(list1)
print("list1 after  sorting :",list1)

```

```python
#Q5 sorting using selection sort
def selSort(A):
    for i in range(len(A)):
        min_idx = i
        for j in range(i+1, len(A)):
            if (A[min_idx] > A[j]):
                min_idx = j
        A[i], A[min_idx] = A[min_idx], A[i]
                
                
list1 = [80,70,90,40,60,50]
print("list1 before sorting :",list1)
selSort(list1)
print("list1 after  sorting :",list1)
```
