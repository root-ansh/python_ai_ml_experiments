Here( in the individual graphs), we plot the KDE graphs of people answering the question about salary (i.e, df_sal_NonNa) and those who answered the question AND have salary less than or equal to USD 250000.  
- The 1st graph show that frequency of popultion that the frequency of population is maximum at the very start of the graph, i.e people having salary between 0 and USD 250K . - The 2nd graph show the frequency of people having salary less than USD 250,000. Here also there seeems to be a large number of population having salary between USD 0-25000.
```python
fig, ax = plt.subplots(figsize=[20, 10])

sub3 = plt.subplot2grid([2, 2], [0, 0])
sns.distplot(df_sal_NonNa['ConvertedSalary'], color='green', kde=True, hist=False, rug=True,
             rug_kws={"color": "green", "alpha": 0.01, "height": 0.02}, ax=sub3, label='df_sal_non_na')

sub4 = plt.subplot2grid([2, 2], [0, 1])
sns.distplot(df_sal_low['ConvertedSalary'], color='orange', kde=True, hist=False, bins=100, rug=True,
             rug_kws={"color": "orange", "alpha": 0.01, "height": 0.02}, ax=sub4, label='df_sal_low')

sub1 = plt.subplot2grid([2, 2], [1, 0], colspan=2)
sns.distplot(df_sal_NonNa['ConvertedSalary'], color='red', kde=True, hist=False, rug=True,
             rug_kws={"color": "red", "alpha": 0.01, "height": 0.02}, ax=sub1, label='df_sal_non_na')

sns.distplot(df_sal_low['ConvertedSalary'], color='blue', kde=True, hist=False, rug=True,
             rug_kws={"color": "blue", "alpha": 0.01, "height": 0.02}, ax=sub1, label='df_sal_low')
```
