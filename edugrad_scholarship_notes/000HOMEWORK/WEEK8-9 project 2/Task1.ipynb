- The Following survey was conducted by 98854 Developers all around the world to collect various trends and information.

###1. Age

The Technological Industry seems to thrive with young and agile Developers as it is found that out of 64573 population , that chose to answer,
49.13% of developers reside between 25-35 years of age.  
This information is crucial and can be further used by government at the time of creation of policies as this data let them know which age group they have to target while devicing Public schemes for this sector.

```python
total_counts = len(df_survey) - 1

all_ages = df_survey.Age.astype('str').value_counts()
not_answered = (all_ages['nan'])
del all_ages['nan']
total_counts = (total_counts - not_answered)

print('total answres =', total_counts)
ax = all_ages.plot(kind='barh', figsize=(10, 6), cmap='rainbow_r')
for i in ax.patches:
    ax.text(
        i.get_width(),
        i.get_y() + 0.3,
        str(round((i.get_width() / total_counts) * 100, 2)) + '% (' + str(i.get_width()) + ')',
        color='black'
    )

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('red')
ax.spines['bottom'].set_color('red')

ax.legend()
ax.grid(which='major', linestyle='--', linewidth='0.5', color='red')

ax.set_title('Devloper Ages Bar Graph.', fontsize=20)
ax.set_xlabel('Age group.', fontsize=12)
ax.set_ylabel('Population.', fontsize=12)
```

###2. Experience
When Analysed ,it is found that 24.8 % of the developers have 3-5 years of coding experience, followed by 20.6% of people with 6-8 years of age.  
This information is crucial and can be further used by companies and job recruiters, since the time spent coding shows how much of knowledge do you have , for a perticular subject

```python
fig = plt.figure(figsize=[13, 6])

all_years_coding = df_survey.YearsCoding.astype('str').value_counts()
del all_years_coding['nan']

ax = all_years_coding.plot(
    kind='pie',
    cmap='jet',
    autopct='%1.1f%%',
    pctdistance=0.8,
    explode=np.random.random(len(all_years_coding)) % 0.2
)
ax.set_title('Coding Experiences Pie Chart.', fontsize=20)
ax.set_ylabel(None)
```

### 3. Gender.  
Gender plays an important role in technological industry. It has been found that 92.29% of  Developers are male, with less than than 7 % of developers being female. the rest of the people being transgender , gender queer and other people giving multiple answers.  
Also, 34,385 people chose not to answer this question, which is arounf 34% of population , when considered. This shows the sensitivity of such question.

```python
total_counts = len(df_survey) - 1

all_genders = df_survey.Gender.astype('str').value_counts()
# not_answered = (all_genders['nan'])
# del all_genders['nan']
# total_counts = (total_counts - not_answered)

print('total answres =', total_counts)
ax = all_genders.plot(kind='barh', figsize=(10, 6), color='#390084')
for i in ax.patches:
    ax.text(
        i.get_width(),
        i.get_y() + 0.3,
        str(round((i.get_width() / total_counts) * 100, 3)) + '% (' + str(i.get_width()) + ')',
        color='black'
    )

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#390084')
ax.spines['bottom'].set_color('#390084')

ax.legend()
ax.grid(which='major', linestyle='--', linewidth='0.5', color='#390084')

ax.set_title('Developer\'s  Gender Bar Graph.', fontsize=20)
ax.set_xlabel('No. of People.', fontsize=12)
ax.set_ylabel('Genders.', fontsize=12)
```

### 4. Country.
Country is an important factor because it can help companies from all over the world to decide which place has the maximum potential to yield developers. Considering the access to the survey to be fair all over the world( since it was an internet based survey, and anyone could participate in it), it was found that USA, India and Germany are the top Countries with largest Developer Population



```python
total_counts = len(df_survey) - 1
all_educations = df_survey.Country.astype('str').value_counts()
not_answered = (all_educations['nan'])
del all_educations['nan']
total_counts = (total_counts - not_answered)

print('total answers =', total_counts)


ax = all_educations.plot(kind='barh', figsize=(10, 60), color='#008429')
for i in ax.patches:
    ax.text(
        i.get_width(),
        i.get_y() + 0.3,
        str(round((i.get_width() / total_counts) * 100, 3)) + '% (' + str(i.get_width()) + ')',
        color='black'
    )

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#008429')
ax.spines['bottom'].set_color('#008429')

ax.legend()
ax.grid(which='major', linestyle='--', linewidth='0.5', color='#008429')

ax.set_title('Developer\'s  Country Bar Graph.', fontsize=20)
ax.set_xlabel('No. of People.', fontsize=12)
ax.set_ylabel('Countries.', fontsize=12)

```
