```python
"""
1). Write a Python program to draw a line with suitable label in the x axis, y axis and a title.
Line function is given as
y = 2 + (3 * x),where x is an integer having values from 1 to 50
"""

x = np.arange(1, 51)
y = x * 2 + 3

plt.plot(x, y, marker='.',label='y= 2x +3')
plt.title('y= 2x+3')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
```

```python
"""
Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 11, 2017.
(I) Create the csv file containing the given Financial data (findata.csv):<in the solution>
(II) Fetch data from the created csv file to create the line plots for different values 'Open', 'High', 'Low' and 'Close' values 
for the given dates.

"""
import csv

l = [
    ('Date', 'Open', 'High', 'Low', 'Close'),
    ('03-10-16', 774.25, 776.065002, 769.5, 772.559998),
    ('04-10-16', 776.030029, 778.710022, 772.890015, 776.429993),
    ('05-10-16', 779.309998, 782.070007, 775.650024, 776.469971),
    ('06-10-16', 779, 780.47998, 775.539978, 776.859985),
    ('07-10-16', 779.659973, 779.659973, 770.75, 775.080017),
    ('10-10-17', 780.382, 777.47998, 776.539978, 776.859985),
    ('11-10-17', 778.659973, 780.659973, 774.75, 780.080017)
]

filename = r'C:\Users\anshs\00ANSH_PYTHON\MyPy\000HOMEWORK\WEEK8\findata.csv'
with open(filename, 'w', newline='') as f:
    csv_writer = csv.writer(f)
    for record in l:
        csv_writer.writerow(record)

print('successfully created csv file')

#part 2
dframe = pd.read_csv(filename)
fig = plt.figure(figsize=[10,10])
plt.subplots_adjust(hspace=0.5)

x_axis=pd.to_datetime(dframe['Date'])

sub1=fig.add_subplot(221)
sub1.tick_params(axis='x',rotation=90)
sub1.set_xlabel("Open")
sub1.plot(x_axis,dframe['Open'],marker='o')


sub2=fig.add_subplot(222)
sub2.tick_params(axis='x',rotation=90)
sub2.set_xlabel("High")
sub2.plot(x_axis,dframe['High'],marker='o')

sub3=fig.add_subplot(223)
sub3.tick_params(axis='x',rotation=90)
sub3.set_xlabel("Low")
sub3.plot(x_axis,dframe['Low'],marker='o')

sub4=fig.add_subplot(224)
sub4.tick_params(axis='x',rotation=90)
sub4.set_xlabel("Close")
sub4.plot(x_axis,dframe['Close'],marker='o')

sub1.grid()
sub2.grid()
sub3.grid()
sub4.grid()
```

```python
"""
4). Write a Python programming to display a bar chart of the popularity of programming Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

>>X- axis should display Languages and Y-axis their Popularity. Mark plot with proper x and y label along with title.
>>X-axis should display ticks with Language names
>>Display Grid on the figure with the following details :- type  major grid, continuous line with width as 1 and red in color
>>Create another plot of the same data but bars should be horizontal and each bar consists of a separate color.
"""
plang = ['Python', 'PHP', 'Java', 'JavaScript', 'C  #', 'C++']
pop = [17.6, 8.8, 22.2, 8, 7.7, 6.7]

fig, sub = plt.subplots()
fig.set_figwidth(6)
fig.set_figheight(6)
fig.suptitle('popularity of programming Languages', fontsize=20)

sub.grid(which='major', color='red', linewidth=1)
sub.set_xlabel('Languages(v)')
sub.set_ylabel('Population(v)')

sub.bar(plang, pop)

sub_new = sub.twiny().twinx()
sub_new.set_xlabel('Population(h)')
sub_new.set_ylabel('Languages(h)')

allbars=sub_new.barh(plang, pop,alpha=0.7)
sub_new.invert_xaxis()

allbars[0].set_color('yellow')
allbars[1].set_color('m')
allbars[2].set_color('orange')
allbars[3].set_color('green')
allbars[4].set_color('blue')
allbars[5].set_color('burlywood')


```

```python
"""
Draw a figure consisting of two circles of diameter 3 units and 4 units respectively using matplotlib.
First circle will be filled with yellow color and the second circle with slanting lines. 
Highlight the center of both with a marker and also annotate with center coordinates having (x,y) values
"""

fig = plt.figure()
sub = fig.add_axes([0, 0, 1, 1])
sub.set_xticks(range(12))
sub.set_yticks(range(12))

sub.set_aspect('equal')

b1 = patches.Circle([4, 4], 1.5, color='yellow')
b2 = patches.Circle([7, 7], 2, hatch='/', edgecolor='white',)

sub.add_patch(b1)
sub.add_patch(b2)

# since the centres are at points (4,4) and (7,7), therefore:
sub.plot([4, 7], [4, 7], marker='o', markerfacecolor='r', markeredgecolor='r', linestyle='None')
sub.annotate('(4,4)', xy=(4, 4),xytext=(4.3, 4))
sub.annotate('(7,7)', xy=(7, 7),xytext=(7.3, 7))

```
