```python
"""
Question:
- In a company named XYZ, employee data is recorded in a spreadsheet called “employee_list.csv”.
- Few more people have joined this company and their data is stored in a file “new_records.csv”.

- This data has to be added in the “employee_list.csv” based on following condition -
    1) The format of “employee_list.csv” is:
            employee code, first name, last name, home city, age, current city
    2)The format of “new_records.csv” is :
            employee code, full name, home city, age, current city
            
    3) The employee code column in “new_records.csv” is prefixed by a three digit 
       code related to type of employee as defined below:
           >> emp – for permanant employees
           >> other than emp – for contractual and other type of employees
           
    4) Add only the permanant employees in the employee_list.csv
    
    5) Create a new separate file for the contractual and other type of employees,
       file name will be “contractors_list.xlsx”
       
    6) the employee code recorded in the employee_list.csv will be only integer, 
       you need to remove chracter prefixed to the employee code.
    
    7)handle the file and number-type exceptions whereever appropriate
 
solution :

code first_name   last_name  home city   age   current city
-----------------------------------------------------------
1    Anshika      Garg       Ranchi       21    Noida
2    Rohit        Awasthi    Allahabad    20    Delhi

new_records.csv:
code   full_name          home_city  age   current_city
-----------------------------------------------------
emp11   Aakash Kumar       Patna      22    Noida
emp12   Naman Arora        Lucknow    18    Delhi
emp13   Vikas Jain         Bhopal     25    Gurgaon
emp14   Nirmal Mishra      Gurgaon    21    Gurgaon
emp15   Priya Singh        Gurgaon    25    Delhi
cnt17   Rishab Srivastava  Mumbai     22    Noida
emp16   Pravita Singh      Kanpur     21    Delhi
tmp18   Naina Gupta        Sonipat    28    New Delhi
cnt19   Jai Tiwari         Hyderabad  24    Noida
 
"""
```

```python
import csv

def createNewRecordsList(filepath,records_list):
    try:
        with open(filepath,'r') as csvFile:
            csv_reader = csv.reader(csvFile)
            for record in csv_reader:
                records_list.append(record)     
    except Exception as e:
        print("System generated the following error:\n\t"+str(e))
        
def splitFullNames(records_list):
    for record in records_list:                         # eg ['emp11', 'Aakash Kumar', 'Patna', '22', 'Noida']
        record[1:2]= record[1].split(' ')              # is now ['emp11', 'Aakash','Kumar', 'Patna', '22', 'Noida']
        
def addToEmployeeList(records_list, filename):
    try:
        with open(filename, 'a',newline='') as f:
            csv_writer = csv.writer(f)
            for record in records_list:
                code=record[0][0:3]  #i.e if record[0]== emp11 , code=emp and ID= 11
                ID= record[0][3:]
                if(code=="emp"):
                    myrec=list(record)
                    myrec[0]=ID
                    csv_writer.writerow(myrec)
    except Exception as e:
        print("System generated the following error:\n\t"+str(e))                   
    
    
def addToContractorsList(records_list):
    filename=r".\1536310594_employee-and-new-record-data\contractors_list.csv"
    try:
        with open(filename, 'a',newline='') as f:   # will create a new file if not lready existing
            csv_writer = csv.writer(f)
            
            for record in records_list:
                code=record[0][0:3]  #i.e if record[0]== emp11 , code=emp
                if(code!="emp"):
                    csv_writer.writerow(record)
    except Exception as e:
        print("System generated the following error:\n\t"+str(e))        
            
    
new_records_list =[]
new_records_filepath=r".\1536310594_employee-and-new-record-data\new_records.csv"
employee_list_fp=r".\1536310594_employee-and-new-record-data\employee_list.csv"

createNewRecordsList(new_records_filepath,new_records_list)  #1) opening file and creating a list of all entries:
splitFullNames(new_records_list)                             #2)splitting the fullname into firstname and second name:
addToEmployeeList(new_records_list,employee_list_fp)         #3)adding records to the employee list(those which are marked 'emp') 
addToContractorsList(new_records_list)



```
